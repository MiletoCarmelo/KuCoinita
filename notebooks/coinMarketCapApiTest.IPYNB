```python
# set up coin market cap api : 
import os
import requests
import json
import pandas as pd
import numpy as np
import datetime
import time
import matplotlib.pyplot as plt

# set up coin market cap api :
cmc_api_key = os.getenv("COINMARKETCAP_API_KEY")
headers = {
  'Accepts': 'application/json',
  'X-CMC_PRO_API_KEY': cmc_api_key,
}

# get the latest cryptocurrency market cap data
def get_latest_cryptocurrency_market_cap():
    url = 'https://pro-api.coinmarketcap.com/v1/cryptocurrency/listings/latest'
    response = requests.get(url, headers=headers)
    data = json.loads(response.text)
    return data

# get the latest cryptocurrency market cap data
def get_historical_cryptocurrency_market_cap(start_date, end_date):
    url = 'https://pro-api.coinmarketcap.com/v1/cryptocurrency/listings/historical'
    parameters = {
      'start': start_date,
      'end': end_date
    }
    response = requests.get(url, headers=headers, params=parameters)
    data = json.loads(response.text)
    return data

def get_exchange_listings(coin_id):
    url = f'https://pro-api.coinmarketcap.com/v1/cryptocurrency/exchange/listings/latest?id={coin_id}'
    response = requests.get(url, headers=headers)
    data = response.json()
    if 'data' in data and data['data']:
      return [exchange['name'] for exchange in data['data']]
    else:
      return None
    
def get_exchanges_id_list():
    url = 'https://pro-api.coinmarketcap.com/v1/exchange/map'
    response = requests.get(url, headers=headers)
    exchanges_map = response.json()
    if 'data' in exchanges_map and exchanges_map['data']:
      return pd.DataFrame(exchanges_map['data'])
    else:
      return None

def get_exchange_assets(exchange_id):
    url = f'https://pro-api.coinmarketcap.com/v1/exchange/assets?id={exchange_id}'
    response = requests.get(url, headers=headers)
    exchange_assets = response.json()
    if 'data' in exchange_assets and exchange_assets['data']:
      return pd.DataFrame(exchange_assets['data'])
    else:
      return None

def get_exchange_info(exchange_id):
    url = f'https://pro-api.coinmarketcap.com/v1/exchange/info?id={exchange_id}'
    response = requests.get(url, headers=headers)
    exchange_info = response.json()
    if 'data' in exchange_info and exchange_info['data']:
      return exchange_info['data']
    else:
      return None

def get_exchange_listings(coin_id):
    url = f'https://pro-api.coinmarketcap.com/v1/cryptocurrency/exchange/listings/latest?id={coin_id}'
    response = requests.get(url, headers=headers)
    data = response.json()
    if 'data' in data and data['data']:
      return [exchange['name'] for exchange in data['data']]
    else:
      return None
```
