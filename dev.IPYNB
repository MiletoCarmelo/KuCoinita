```python
# libraries used : 
# 1) requests 
# 2) time
# 3) pandas
# 4) openpyxl
```

```python
# get the unixtime 
import time
unixtime = int(time.time())
```

```python
# create api request : 
req = "https://api.kucoin.com/api/v1/market/allTickers?timestamp={}".format(unixtime)
```

```python
# request the api
import requests
r = requests.get(req)
# convert to dataframe
import pandas as pd
df = pd.DataFrame(r.json()['data']['ticker'])
# get time stamp 
df['timestamp'] = r.json()['data']['time']
# arrange by vol desc 
df = df.sort_values(by='vol', ascending=False)
df
```

```python
# get date from timestamp column
from datetime import datetime
df['date'] = df['timestamp'].apply(lambda x: datetime.fromtimestamp(x/1000.0).strftime('%Y-%m-%d'))
```

```python
time_label = df['date'].unique()[0]
time_label
```

```python
# export to an excel file in a new sheet (if the excel file does not exist, it will be created):
from openpyxl import load_workbook
# check if file exists
import os.path
if os.path.isfile('kucoin.xlsx') == False:
    df.to_excel('kucoin.xlsx', sheet_name='{}'.format(time_label), index=False)
else: 
    book = load_workbook('kucoin.xlsx')
    writer = pd.ExcelWriter('kucoin.xlsx', engine='openpyxl')
    writer.book = book
    writer.sheets = dict((ws.title, ws) for ws in book.worksheets)
    df.to_excel(writer, '{}'.format(time_label))
    writer.save()
    writer.close()
```

```python
# export to google drive : 
from pydrive.auth import GoogleAuth
from pydrive.drive import GoogleDrive
gauth = GoogleAuth()
gauth.LocalWebserverAuth()
drive = GoogleDrive(gauth)
file = drive.CreateFile()
file.SetContentFile('kucoin.xlsx')
file.Upload()
file['title'] = 'kucoin.xlsx'
file.Upload()
file.InsertPermission({
    'type': 'anyone',
    'value': 'anyone',
    'role': 'reader'
})
print('File uploaded to Google Drive')
```
